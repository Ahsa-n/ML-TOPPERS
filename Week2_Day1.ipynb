```python
url = 'https://raw.githubusercontent.com/Ahsa-n/ML-TOPPERS/refs/heads/main/Week%201/project_scores.csv'
df = pd.read_csv(url)
df.head(3)
```

```python
df.loc[:,['ML_Score','Age']]
```

```python
df['Total_Score']= df.ML_Score + df.Python_Score + df.AI_Score
df
```

```python
df =df.fillna(0)
df['Total_Score']= df.ML_Score + df.Python_Score + df.AI_Score
df
```

```python
import numpy as np
df['results'] =np.where(df['Total_Score'] >= 240,'pass','fail' )
df
```
