```python
import numpy as np

sales_data = np.array([
    [200, 220, 250, 210, 180, 190],  # Day 1
    [230, 240, 260, 200, 195, 205],  # Day 2
    [210, 215, 255, 220, 185, 200],  # Day 3
    [205, 225, 270, 215, 190, 195],  # Day 4
    [215, 230, 265, 225, 200, 210],  # Day 5
    [225, 235, 275, 230, 205, 215],  # Day 6
    [235, 245, 280, 240, 210, 220],  # Day 7
    [245, 255, 290, 250, 215, 225],  # Day 8
    [255, 265, 300, 260, 220, 230],  # Day 9
    [265, 275, 310, 270, 225, 235],  # Day 10
])
#q1 a)
totalsaleperproduct = np.sum(sales_data,axis=0)
for i,x in enumerate(totalsaleperproduct):
  print(f"total sale of product no {i+1}: {x}")
#b)
for i,x in enumerate(totalsaleperproduct):
  print(f"average daily sale of product no {i+1}: {x/10:.2f}")
#c)
newsales_data= sales_data * 1.05
print(newsales_data)
#d)
newsales_data = np.sqrt(sales_data)
print(np.round(newsales_data,2))

#q2 a)
bonus_array = np.array([10,20,15,25,30,5])
newsales_data = sales_data + bonus_array
print(np.round(newsales_data,2))
#b)
newsales_data= sales_data + 50
print(np.round(newsales_data,2))
#q3 a)
print(f"mean: {np.mean(sales_data)}\nmedian: {np.median(sales_data)}\nvariance: {np.var(sales_data)}\nsd: {np.std(sales_data)} ")
#b)
print(f"max = {np.max(sales_data)}\nmin = {np.min(sales_data)}\nrange = {np.max(sales_data) - np.min(sales_data)}")
#c)
q1 = np.percentile(sales_data,25)
q3 = np.percentile(sales_data,75)
iqr = q3-q1
print(f"iqr: {iqr}")
#q4 a)
newarr = sales_data > 250
newsales=  sales_data[newarr]
print(newsales)
#b)
newarr = sales_data>300
sales_data[newarr] = 300
print(sales_data[newarr])
#c)
newarr = (sales_data>= 200) & (sales_data<= 250)
print(np.count_nonzero(sales_data[newarr]))
#d)
newarr = sales_data < np.mean(sales_data)
modifiedarr = sales_data[newarr]
print(modifiedarr)
print(sales_data)
print(np.mean(sales_data))
#q5
#a)
print(np.sort(sales_data[4]))
#b)
sumarr= np.sum(sales_data,axis = 1)
print(f"day: {np.argmax(sumarr) +1}")
#c)
print(np.mean(sales_data,axis = 0))
#d)
print(np.round(np.mean(sales_data,axis =1),2))
#e)
overall_average = np.mean(sales_data)
print(f"Overall average sales: {overall_average:.2f}")
#q6
def highlight_outliers(data):
  tempmean = np.mean(data)
  sd = np.std(data)
  outlier = (tempmean + 2*sd >data) & (tempmean- 2*sd <data)
  return data[outlier]
print(highlight_outliers(sales_data))
```
